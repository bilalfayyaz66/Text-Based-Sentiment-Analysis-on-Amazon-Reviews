```python
def rule_sentiment(text):
    neg_patterns = [r"\b(?:bad|poor|negative)\b", r"\b(?:unhappy|sad)\b"]
    pos_patterns = [r"\b(?:good|excellent|positive)\b", r"\b(?:happy|joyful)\b"]
    

    for pattern in neg_patterns:
        if re.search(pattern, text, flags=re.IGNORECASE):
            return 'Negative'
    for pattern in pos_patterns:
        if re.search(pattern, text, flags=re.IGNORECASE):
            return 'Positive'
    return 'Neutral'

def get_sentiment(reviews, threshold=3.0, num_phrases=20):
    pos_phrases = []
    neg_phrases = []
    
    for review_text in reviews:
        withoutpunc = remove_punctuation(review_text)
        withoutstopwords= remove_stop_words(withoutpunc)
        sentiment = rule_sentiment(withoutstopwords)

        if sentiment == 'Positive':
            pos_phrases.extend(withoutstopwords.lower().split())
        elif sentiment == 'Negative':
            neg_phrases.extend(withoutstopwords.lower().split())

    pos_com = [phrase for phrase, _ in Counter(pos_phrases).most_common(num_phrases)]
    neg_com = [phrase for phrase, _ in Counter(neg_phrases).most_common(num_phrases)]

    return pos_com, neg_com
threshold = 3.0
# randomly
preprocessed_data = [
    "The product is really good and I'm happy with it.",
    "I'm really unhappy with this purchase. It's bad.",
    "The service was excellent. I'm very pleased.",
    "I'm sad because the item arrived damaged."
]
pos_phrases, neg_phrases = get_sentiment(preprocessed_data, threshold)
print("\nNegative phrases:")
print(neg_phrases)

print("Positive phrases:")
print(pos_phrases)



#by getting preprocessed data
preprocessed_data = [preprocessed_data]
pos_phrases, neg_phrases = get_sentiment(preprocessed_data, threshold)
#printing outputs
print("\Positive phrases:")
print(pos_phrases)

print("\nNegative phrases:")
print(neg_phrases)

print(end='\n')

#in semantic analysis, semantic score calculationss functions
def sentiment_score(text, weights):
    withoutpunc = remove_punctuation(text)
    words = withoutpunc.lower().split()
    sentiments_score = sum(weights.get(word, 0) for word in words)
    return sentiments_score
#classifying 
def classify_sentiment(score, threshold=0.5):
    #conditions
    if score > threshold:
        return 'Positive'
    elif score < -threshold:
        return 'Negative'
    else:
        return 'Neutral'
#getsentiment function with attriutes
def get_sentiment(reviews, word_weights, threshold=0.5, num_phrases=20):
    pos_phrases = []
    neg_phrases = []
#conditionsss
    for review_text in reviews:
        if isinstance(review_text, str):  # Ensure review_text is a string
            sentiments_score = sentiment_score(review_text, word_weights)
            sentiment = classify_sentiment(sentiments_score, threshold)

            if sentiment == 'Negative':
                neg_phrases.extend(review_text.lower().split())
            elif sentiment == 'Positive':
                pos_phrases.extend(review_text.lower().split())

            print(f"Review: {review_text[:50]}... | Sentiment: {sentiment} | Score: {sentiments_score}")

    neg_common = [phrase for phrase, _ in Counter(neg_phrases).most_common(num_phrases)]
    pos_common = [phrase for phrase, _ in Counter(pos_phrases).most_common(num_phrases)]
    
    return pos_common, neg_common
#defining weightss
weights = {'good': 0.7, 'amazing': 0.8, 'bad': -0.05, 'poor': -0.5, 'excellent': 0.9}
threshold = 0.5

# random
preprocessed_data = [ "The product is really good and I'm happy with it.",
    "I'm really unhappy with this purchase. It's bad.",
    "The service was excellent. I'm very pleased.",
    "I'm sad because the item arrived damaged."]

pos_phrases, neg_phrases = get_sentiment(preprocessed_data, weights, threshold)
print("Positive Phrases:")
print(pos_phrases)
print("\nNegative Phrases:")
print(neg_phrases)
```
